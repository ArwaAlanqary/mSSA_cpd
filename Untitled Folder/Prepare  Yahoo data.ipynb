```python
# get yahoo_files
files = glob.glob('../data/yahoo/raw/A4Benchmark/A4Benchmark-TS**.csv')
for file in files:
    # load yahoo data
    df = pd.read_csv(file)
    # create ts df
    df_ts = pd.DataFrame()
    df_ts['v'] = df['value']
    ts_name = file.split('/')[-1][:-4]
    df_ts.to_csv(f'../data/yahoo/{ts_name}_ts.csv', header = None)
    # create labels df
    df_labels = pd.DataFrame(columns = [0,1])
    cpts = np.nonzero(df['changepoint'].values)[0]
    start = 0 
    for i,pt in enumerate(cpts):
        df_labels.loc[i] = [start, pt-1]
        start = pt
    df_labels.to_csv(f'../data/yahoo/{ts_name}_labels.csv', header = None, index= False)

  
```
